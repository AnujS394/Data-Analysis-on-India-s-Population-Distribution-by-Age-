```python
import matplotlib.pyplot as plt

# Example gender distribution (replace with real data)
gender_data = {'Male': 480, 'Female': 520}

# Bar chart
plt.figure(figsize=(6, 4))
plt.bar(gender_data.keys(), gender_data.values(), color=['skyblue', 'lightpink'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Number of People')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example age data (replace with your dataset)
np.random.seed(0)
age_data = np.random.normal(loc=35, scale=10, size=1000)  # 1000 people with average age 35

# Histogram
plt.figure(figsize=(8, 5))
plt.hist(age_data, bins=20, color='lightgreen', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
